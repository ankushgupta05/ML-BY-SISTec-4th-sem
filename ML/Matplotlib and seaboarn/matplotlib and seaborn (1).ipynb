# scatter plot

```python
plt.scatter(x,y,c='g')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('graph in 2d')
plt.savefig('test.png')
```

# line plot

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=4,markersize=20)
plt.xlabel('x value')
plt.ylabel('v value')
plt.title('2d dig')
```

# subplot 

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.savefig('subdig')
```

# pieplot

```python
labels='python','c++','java','c'
size=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0,0,0,0.2)
plt.pie(size,labels=labels,colors=colors,autopct="%1.1f%%",explode=explode)
```

```python
labels='python','c++','java','c'
size=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0)
plt.pie(size,labels=labels,colors=colors,autopct='%1.1f%%',explode=explode)
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
#correlation
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#pairplot
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```
