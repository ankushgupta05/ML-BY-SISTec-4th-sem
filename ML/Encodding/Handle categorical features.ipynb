# task  convert Embark colum in numerical by on hot encodding

```python
df2 = df[['Embarked']]
df2
```

```python
df2['Embarked'].unique()
```

```python
dummies = pd.get_dummies(df2)
dummies
```

```python
df3 = pd.concat([df,dummies], axis=1)   # concat dummies columns and df dataset
df3.head()
```

```python
df4 = df3.drop(['Embarked'], axis=1)
df4.head()   # delete Embarked columns
```
