# ONE HOTE ENCODDING WITH SKLEARN

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
rd = one_h.fit_transform(data[['Genre']])   # it give o/p in a numpy array
rd
```

```python
data2 = pd.DataFrame(rd,columns=['male','female'])
data2
```

```python
df = pd.concat([data,data2], axis=1)
df
```

# Label Encodding

```python
df = pd.read_csv('sample data\\name.csv')
df
```

```python
df2 =df[['size']]   # it give dataframe
df2



# df2 =df[['size']]   # it give numpy array
# df2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le
```

```python
print(df2)

print()
print()

le.fit_transform(df2)
```

```python
df2['size_update'] = le.fit_transform(df2)
df2
```

```python
df['size'] = df2['size_update']
df
```
